```python
import tensorflow as tf
from tensorflow.keras import layers
```

```python
def df_to_dataset(dataframe, shuffle=True, batch_size=32):
    df = dataframe.copy()
    labels = df.pop('target')
    df = {key: value[:,tf.newaxis] for key, value in dataframe.items()}
    ds = tf.data.Dataset.from_tensor_slices((dict(df), labels))
    if shuffle:
        ds = ds.shuffle(buffer_size=len(dataframe))
    ds = ds.batch(batch_size)
    ds = ds.prefetch(batch_size)
    return ds
```

```python
batch_size = 5
train_ds = df_to_dataset(df, batch_size=batch_size)
```

```python
def get_normalization_layer(name, dataset):
    # Create a Normalization layer for the feature.
    normalizer = layers.Normalization(axis=None)

    # Prepare a Dataset that only yields the feature.
    feature_ds = dataset.map(lambda x, y: x[name])

    # Learn the statistics of the data.
    normalizer.adapt(feature_ds)

    return normalizer
```

```python
photo_count_col = train_features['Team0Ele0']
layer = get_normalization_layer('Team0Ele0', train_ds)
layer(photo_count_col)
```

```python
def get_category_encoding_layer(name, dataset, dtype, max_tokens=None):
    # Create a layer that turns strings into integer indices.
    if dtype == 'string':
        index = layers.StringLookup(max_tokens=max_tokens)
    # Otherwise, create a layer that turns integer values into integer indices.
    else:
        index = layers.IntegerLookup(max_tokens=max_tokens)

    # Prepare a `tf.data.Dataset` that only yields the feature.
    feature_ds = dataset.map(lambda x, y: x[name])

    # Learn the set of possible values and assign them a fixed integer index.
    index.adapt(feature_ds)

    # Encode the integer indices.
    encoder = layers.CategoryEncoding(num_tokens=index.vocabulary_size())

    # Apply multi-hot encoding to the indices. The lambda function captures the
    # layer, so you can use them, or include them in the Keras Functional model later.
    return lambda feature: encoder(index(feature))
```

```python
def create_cnn(width, height, depth, filters=(16, 32, 64)):
    skill_input = Input(shape=(2,))
    
    # initialize the input shape and channel dimension, assuming
    # TensorFlow/channels-last ordering
    inputShape = (width, height, depth)
    chanDim = -1
    # define the model input
    inputs = Input(shape=inputShape)
    # loop over the number of filters
    for (i, f) in enumerate(filters):
        # if this is the first CONV layer then set the input
        # appropriately
        if i == 0:
            x = inputs
        # CONV => RELU => BN => POOL
        x = Conv2D(f, (3, 3), padding="same", activation='relu')(x)
        x = BatchNormalization(axis=chanDim)(x)
        x = MaxPooling2D(pool_size=(2, 2))(x)
        
        # flatten the volume, then FC => RELU => BN => DROPOUT
    x = Flatten()(x)
    x = Concatenate()([x, skill_input])
    x = Dense(16, activation='relu')(x)
    x = BatchNormalization(axis=chanDim)(x)
    x = Dropout(0.5)(x)
    # apply another FC layer, this one to match the number of nodes
    # coming out of the MLP
    x = Dense(4, activation='relu')(x)
    # regression output
    x = Dense(1, activation="linear")(x)
    # construct the CNN
    model = Model(inputs=[inputs, skill_input], outputs=x)
    # return the CNN
    return model
```

```python
model = create_cnn(50, 49, 1)
opt = Adam(lr=1e-3, decay=1e-3 / 200)
model.compile(loss="mean_absolute_percentage_error", optimizer=opt)
```

```python
from keras.utils.vis_utils import plot_model
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
# scale pixels
def prep_pixels(train, test):
    # convert from integers to floats
    train_norm = train.astype('float32')
    test_norm = test.astype('float32')
    # normalize to range 0-1
    train_norm = train_norm / 5
    test_norm = test_norm / 5
    # return normalized images
    train_norm = train_norm.reshape((train_norm.shape[0], train_norm.shape[1], train_norm.shape[2], 1))
    test_norm = test_norm.reshape((test_norm.shape[0], test_norm.shape[1], test_norm.shape[2], 1))
    return train_norm, test_norm
```

```python
train_x, val_x = prep_pixels(train_x, val_x)
```
