```python
MAX_VALUE = 2100
BASE_VALUE = 1200
def normalize_elo(elo):
    return (elo - BASE_VALUE) / (MAX_VALUE / BASE_VALUE)
```

```python
def one_hot_encode(name):
    data = {
        "AOE": [0, 0, 0, 1],
        "Tank": [0, 0, 1, 0],
        "Base": [0, 1, 0, 0],
        "Healer": [1, 0, 0, 0]
    }
    return data[name]
```

```python
raw_data = {
    "Team0Character0": one_hot_encode("AOE"),
    "Team0Ele0": normalize_elo(1800),
    "Team0Character1": one_hot_encode("AOE"),
    "Team0Ele1": normalize_elo(1900),
    "Team1Character0": one_hot_encode("AOE"),
    "Team1Ele0": normalize_elo(1600),
    "Team1Character1": one_hot_encode("AOE"),
    "Team1Ele1": normalize_elo(1200),
    "target": 4
}

data = []

for k in raw_data.keys():
    values = raw_data[k]
    if isinstance(values, list):
        for v in values:
            data.append(v)
    else:
        data.append(values)

print(data)
```

```python
def create_cnn(width, height, depth, filters=(16, 32, 64)):
    skill_input = Input(shape=(2,))
    
    # initialize the input shape and channel dimension, assuming
    # TensorFlow/channels-last ordering
    inputShape = (width, height, depth)
    chanDim = -1
    # define the model input
    inputs = Input(shape=inputShape)
    # loop over the number of filters
    for (i, f) in enumerate(filters):
        # if this is the first CONV layer then set the input
        # appropriately
        if i == 0:
            x = inputs
        # CONV => RELU => BN => POOL
        x = Conv2D(f, (3, 3), padding="same", activation='relu')(x)
        x = BatchNormalization(axis=chanDim)(x)
        x = MaxPooling2D(pool_size=(2, 2))(x)
        
        # flatten the volume, then FC => RELU => BN => DROPOUT
    x = Flatten()(x)
    x = Concatenate()([x, skill_input])
    x = Dense(16, activation='relu')(x)
    x = BatchNormalization(axis=chanDim)(x)
    x = Dropout(0.5)(x)
    # apply another FC layer, this one to match the number of nodes
    # coming out of the MLP
    x = Dense(4, activation='relu')(x)
    # regression output
    x = Dense(1, activation="linear")(x)
    # construct the CNN
    model = Model(inputs=[inputs, skill_input], outputs=x)
    # return the CNN
    return model
```

```python
model = create_cnn(50, 49, 1)
opt = Adam(lr=1e-3, decay=1e-3 / 200)
model.compile(loss="mean_absolute_percentage_error", optimizer=opt)
```

```python
# scale pixels
def prep_pixels(train, test):
    # convert from integers to floats
    train_norm = train.astype('float32')
    test_norm = test.astype('float32')
    # normalize to range 0-1
    train_norm = train_norm / 5
    test_norm = test_norm / 5
    # return normalized images
    train_norm = train_norm.reshape((train_norm.shape[0], train_norm.shape[1], train_norm.shape[2], 1))
    test_norm = test_norm.reshape((test_norm.shape[0], test_norm.shape[1], test_norm.shape[2], 1))
    return train_norm, test_norm
```
